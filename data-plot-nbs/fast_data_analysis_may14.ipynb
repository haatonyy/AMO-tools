```python
import numpy as np
import matplotlib.pyplot as plt


import sys
from pathlib import Path

PROJECT_ROOT = Path.cwd().parent  # if notebook is in AMO-tools/notebooks
sys.path.insert(0, str(PROJECT_ROOT))

from exp.data_analysis import load_data, load_param, fit_linear_background_iterative
from exp.science import laser_current_to_freq_MHz_2
from glob import glob
```

```python
data_folder = "../data/data-experiment/05142026/05142026-235743/"
data_file = glob(data_folder+"*.csv")[0]
data = load_data(data_file)
```

```python
i_mA =  data["current_mA"][20:100]
err = data["ch1_V"][20:100]
value = data["ch3_V"][20:100]
```

```python
import matplotlib.pyplot as plt
import numpy as np


fig, (ax1, ax2) = plt.subplots(
    2, 1,          # 2 rows, 1 column
    sharex=True,   # same x-axis
    figsize=(6, 5)
)

# Top plot
ax1.plot(i_mA, value)
ax1.set_ylabel("V")
ax1.vlines(72.64, 0.17,0.19, color="red")

# Bottom plot
ax2.plot(i_mA, err)
ax2.set_ylabel("V")
ax2.set_xlabel("mA")
ax2.vlines(72.62, -0.02,0.02, color="red")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np



fig, ax1 = plt.subplots()

# First y-axis
ax1.plot(i_mA, value)
ax1.set_xlabel('I (mA)')
ax1.set_ylabel('V')

# Second y-axis sharing same x-axis
ax2 = ax1.twinx()
ax2.plot(i_mA, err, '--', label='cos(x)', color="red")
ax2.set_ylabel('V')
ax2.hlines(0, 72.2, 73, color="black")

plt.show()
```
