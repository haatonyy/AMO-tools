```python
import serial
import time
def write_value(ser, command_root, value, print_setup=False):
    command = command_root + bytes(" {value}\n".format(value=value), "utf-8")
    #print(command)
    ser.write(command)
    ser.read(len(command)-1)
    if print_setup==True:
        ser.write(command_root+b"?\n")
        print(ser.readline())
```
